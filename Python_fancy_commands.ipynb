```python
#@title ‚è±Ô∏è %%timeit for basic optimizations
#@markdown **n** - number of runs within a round
#@markdown <br>**r** - number of rounds

%%timeit -n 500000 -r 10

def deduplicate_string(string):
    return list(dict.fromkeys(string))

def merge_the_tools(string, k):
    splits = len(string) // k
    split_strings = [deduplicate_string(string[i*k:(i+1)*k]) for i in range(splits)]

    for substring in split_strings:
        return ''.join(substring)

string = 'AABCAAADA'
k=3

merge_the_tools(string, k)
```

```python
#@title üíæ Store %timeit into a variable

result = %timeit -o merge_the_tools(string,k)
```
